```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

file_path = "C:/Users/bodic/Downloads/cytof_one_experiment.csv"
df = pd.read_csv(file_path)

df.info(), df.head()
```

```python
# Select the column for distribution analysis
column = "GranzymeB"

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
sns.histplot(df[column], bins=50, kde=False, ax=axes[0], color="skyblue")
axes[0].set_title(f"Histogram of {column}")
axes[0].set_xlabel(column)
axes[0].set_ylabel("Frequency")

# Density Estimate (KDE)
sns.kdeplot(df[column], ax=axes[1], color="darkblue")
axes[1].set_title(f"Density Estimate of {column}")
axes[1].set_xlabel(column)
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

**Interpretation of the Distribution Plots:**

**Histogram:** The histogram of "GranzymeB" shows the overall distribution of values, with a peak in a certain range. The shape indicates whether the data is skewed or normally distributed.

**Density Estimate (KDE):** The KDE plot provides a smoothed version of the histogram, showing the probability density. It helps identify if the distribution is unimodal or multimodal.

**Analysis:** The graphs shows a negative skew.

```python
# Select columns for Q-Q plot comparison
col1 = "GranzymeB"
col2 = "Perforin"

fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(df[col1], dist="norm", plot=ax)
stats.probplot(df[col2], dist="norm", plot=ax)

ax.legend([col1, col2], loc="best")
ax.set_title(f"Q-Q Plot: {col1} vs {col2}")
plt.show()
```

**Interpretation of the Q-Q Plot:**

Both columns exhibit deviations, suggesting they may not be perfectly normal.

Differences in curvature or spread indicate variations in distribution shape.
